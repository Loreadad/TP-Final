```python
from calendar import c
import requests
import datetime
import sqlite3 as sql
import pandas as pd
import csv


com = input("Ingrese ticket: ")

startTime= input("Ingrese fecha desde: ") 
año, mes, dia = map(int, startTime.split('-'))


endtime= input("Ingrese fecha hasta: ")
año, mes, dia = map(int, endtime.split('-'))

fecha = startTime + "/" + endtime

url = "https://api.polygon.io/v2/aggs/ticker/"+ str(com) +"/range/1/day/"+ str(fecha)+ "?adjusted=true&sort=asc&limit=120&apiKey=RInphImvZJ9p3rbDMZF76xr712N4Abhs"

payload={}
files={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload, files=files)

res = []
data = response.json()

for s in data['results']:
    res.append(s)

df = pd.json_normalize (res)
df.to_csv ('archivo_graficar.csv')

filename = 'archivo_graficar.csv'
file =open(filename, 'r')
reader = csv.reader(file, delimiter=',')

sql = sql.connect('archivo_graficar.db')
cur = sql.cursor()

#Creamos la tabla si no existe
cur.execute('''CREATE TABLE IF NOT EXISTS archivo_graficar
            (v integer, vw integer, o float, c float, h float, l float, t integer, n integer)''')

#Llenamos la BD con los datos del CSV
for row in reader:
    cur.execute("INSERT INTO archivo_graficar VALUES (?, ?, ?, ?, ?, ?, ?, ?)", (row[0], row[1], row[2], row[3],row[4],row[5],row[6],row[7]))

#Muestro las filas guardadas en la tabla
for row in cur.execute('SELECT * FROM archivo_graficar'):
    print(row)

#Cerramos el archivo y la conexion a la bd
file.close()
sql.commit()
sql.close()

```
